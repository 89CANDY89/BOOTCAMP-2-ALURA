# Setup

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
pd.options.display.float_format = '{:.2f}'.format
```

```python
dados = pd.read_csv("https://raw.githubusercontent.com/89CANDY89/BOOTCAMP-2-ALURA/main/dados/COVID%20DADOS.csv", encoding='ISO-8859-1', skiprows=0, sep = ';', skipfooter=12, thousands='.', decimal=',', engine='python')
```

## ANALISE

```python
usaveis = dados.mean().index.tolist()
usaveis.insert(0, 'data')
dados_usaveis = dados[usaveis]
dados_usaveis = dados_usaveis.set_index('data')
dados_usaveis = dados_usaveis.drop(['populacaoTCU2019','coduf', 'interior/metropolitana','semanaEpi','casosNovos', 'obitosAcumulado',
       'obitosNovos','emAcompanhamentoNovos'], axis=1)


obitos = dados.mean().index.tolist()
obitos.insert(0, 'data')
dados_obitos = dados[obitos]
dados_obitos = dados_obitos.set_index('data')
dados_obitos = dados_obitos.drop(['populacaoTCU2019','coduf', 'interior/metropolitana','semanaEpi', 'casosAcumulado','Recuperadosnovos', 'emAcompanhamentoNovos', 'obitosAcumulado'], axis=1)
```

```python
dados_usaveis.columns
```

```python
dados_obitos.columns
```

```python
fig, axs = plt.subplots(1,2, figsize=(20,8))

dados_obitos.plot(grid=True, ax=axs[1])

dados_usaveis.plot(grid=True, ax=axs[0])

axs[0].set_title('CASOS ACUMULADOS E RECUPERADOS')
axs[0].set_xlabel('DATA')
axs[0].set_ylabel('QUANTIDADE DE PESSOAS')

axs[1].set_title('OBITOS ACUMULADOS E NOVOS CASOS')
axs[1].set_xlabel('DATA')
axs[1].set_ylabel('QUANTIDADE DE PESSOAS')

plt.ylim(3,8)
plt.suptitle('GRAFICOS DE CASOS ACUMULADOS E DE OBITOS', fontsize=18)

plt.show()
```
